```python
import pandas as pd
import matplotlib.pyplot as plt

# load results/results_1cd2.csv with pandas
df_train = pd.read_csv('results/results_train_1cd2.csv')
df_test = pd.read_csv('results/results_val_1cd2.csv')
```

```python
#plot the loss of train and test
plt.figure(figsize=(5, 5))
plt.plot(df_train['loss'], label='train')
plt.plot(df_test['loss'], label='val')
plt.title('Loss over epochs - epochs=25, lr=0.1, batch_size=128')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# plot accuracy of train and test
plt.figure(figsize=(5, 5))
plt.plot(df_train['top1_acc'], label='train')
plt.plot(df_test['top1_acc'], label='val')
plt.title('Accuracy over epochs - epochs=25, lr=0.1, batch_size=128')
plt.xlabel('Epoch')
plt.ylabel('Accuracy [%]')
plt.legend()
plt.show()
```
